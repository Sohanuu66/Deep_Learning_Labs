```python
import pandas as pd
import numpy as np
```

```python
def sigmoid(z):
  return 1/(1 + np.exp(-z))

def delta_sigmoid(z):
  temp = sigmoid(z)
  return temp * (1 - temp)

def step(z):
    return (z >= 0.5).astype(int)
```

```python
class Layer:
    def __init__(self, W):
        self.W = W

    def forward(self, X):
        self.X = X
        self.Z = self.X @ self.W.T
        self.A = sigmoid(self.Z)
        return self.A

    def backward(self, dA, alpha):
        dZ = dA * delta_sigmoid(self.Z)
        dW = dZ.T @ self.X
        dX = dZ @ self.W

        self.W -= alpha * dW
        return dX
```

```python
X = np.array([[1, 0.7, 1.2]])
y = np.array([[1, 0]])

W1 = np.array([[0.5, 1.5, 0.8],
               [0.8, 0.2, -1.6]])

W2 = np.array([[0.9, -1.7, 1.6],
               [1.2,  2.1, -0.2]])

hidden_layer = Layer(W1)
output_layer = Layer(W2)
```

```python
epochs = 1000
alpha = 0.1
theta = 0.5

for epoch in range(epochs):

    # Forward
    h = hidden_layer.forward(X)
    h_bias = np.concatenate([np.ones((1,1)), h], axis=1)
    y_pred = output_layer.forward(h_bias)

    # Loss
    loss = 0.5 * np.sum((y - y_pred) ** 2)

    # Backward
    dL_dy = y_pred - y
    d_hidden_bias = output_layer.backward(dL_dy, alpha)
    d_hidden = d_hidden_bias[:, 1:]
    hidden_layer.backward(d_hidden, alpha)

    if epoch % 100 == 0:
        print(f"Prediction {y_pred}, Epoch {epoch}, Loss: {loss:.6f}")

print("\nFinal continuous output:", y_pred)
print("Final thresholded output:", step(y_pred))
```
